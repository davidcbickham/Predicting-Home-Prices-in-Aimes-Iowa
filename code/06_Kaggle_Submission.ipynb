# Kaggle Submission

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import pickle
import numpy as np
np.random.seed(42)
```

## Load in trained model and scaler

```python
!pwd
```

```python
with open('../assets/scaler_1553209257.pkl', 'rb') as f:
    ss = pickle.load(f)
    
with open('../assets/lin_reg.pkl', 'rb') as f:
    lr = pickle.load(f)
    
with open('../assets/columns_1553209257.pkl', 'rb') as f:
    cols = pickle.load(f)
```

## Load kaggle data

```python
kaggle = pd.read_csv('../data/test.csv', index_col = 'Id',na_values='', keep_default_na=False)
```

```python
kaggle.head()
```

## Drop Fields and Handle Case

```python
kaggle.columns = [x.lower().replace(' ','_') for x in kaggle.columns]
```

```python
kaggle.drop('pid', axis=1,inplace=True)
```

```python
kaggle.drop('misc_val',axis=1,inplace=True)
```

## Clean data in same method as training data

```python
kaggle.fillna(
{'mas_vnr_type': 'None'
 ,'mas_vnr_area': 0.0
 ,'bsmt_exposure': 'No'
  ,'bsmt_qual': 'TA'
 ,'bsmt_cond': 'TA'
    ,'bsmtfin_type_1': 'GLQ'
    ,'bsmtfin_type_2': 'Unf'
    ,'garage_finish': 'Unf'
    ,'garage_qual': 'TA'
    ,'garage_cond': 'TA'
     ,'lot_frontage': 69.0
     ,'garage_yr_blt': 1978
     ,'bsmt_half_bath': 0.0
     ,'bsmt_full_bath': 0.0
     ,'garage_area': 0.0
     ,'garage_cars': 0.0
     ,'bsmtfin_sf_1': 368.0
     ,'bsmt_unf_sf': 0.0
     ,'bsmtfin_sf_2': 0.0
     ,'electrical': 'SBrkr'
}
,inplace=True)
```

```python
kaggle.isna().sum().sum()
```

```python
kaggle.shape
```

## Get Dummies

```python
test_kaggle = pd.get_dummies(kaggle)
```

```python
test_kaggle.shape
```

```python
set(cols).difference(test_kaggle.columns)
```

```python
test_kaggle[cols].head()
```

```python
test_kaggle['misc_val'] = 0
```

## Scale data

```python
kaggle_sc = ss.transform(test_kaggle[cols])
```

## Make predictions

```python
preds = lr.predict(kaggle_sc)
```

## Align predictions back with index and set up header

```python
submission = pd.DataFrame(preds, index=kaggle.index, columns=['SalePrice'])
```

## Sort index (required for proper submission)

```python
submission.sort_index(inplace=True)
```

## Save to csv to submit

```python
submission.to_csv('../data/numeric_only_lr.csv')
```
